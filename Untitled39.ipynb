```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#loading the datasets we will use
clubGames_data = pd.read_csv('/content/club_games.csv')
clubs_data = pd.read_csv('/content/clubs.csv')
```

```python
clubs_data.head()
```

```python
clubGames_data.head()
```

```python
# Perform the integration based on the common columns
merged_data = pd.merge(clubGames_data, clubs_data, on='club_id', how='left')

#saving the new data into a new csv file
merged_data.to_csv('/content/merged_dataset.csv', index=False)
```

```python
merged_data.head()
```

```python
#first question: the relationship between the average age of a football club's squad and the number of own goals scored in games?

#group the data by the club_id
grouped_data = merged_data.groupby('club_id')

#calculating the average age and the total own goals
club_stats = grouped_data.agg({'average_age': 'mean', 'own_goals': 'sum'}).reset_index()

#Analyze the correlation
correlation = club_stats['average_age'].corr(club_stats['own_goals'])


print(f"Correlation between Average Age and Total Goals: {correlation}")

#Visualization
plt.scatter(club_stats['average_age'], club_stats['own_goals'])
plt.xlabel('Average Age of Squad')
plt.ylabel('Total Goals Scored')
plt.title('Relationship between Average Age and Total Goals')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='club_id', y='average_age', data=club_stats, color='blue', label='Average Age')
sns.barplot(x='club_id', y='own_goals', data=club_stats, color='red', label='Total Own Goals')
plt.xlabel('Football Club ID')
plt.ylabel('Values')
plt.title('Comparison of Average Age and Total Own Goals for Different Clubs')
plt.legend()
plt.show()
```

```python
#second question:  Is there a correlation between a football club hosting a game and the likelihood of winning?

# Group the data by club and calculate the total number of home games and wins
grouped_data = merged_data.groupby('club_id')
club_stats_new = grouped_data.agg({'hosting': lambda x: (x == 'Home').sum(), 'is_win': 'sum'}).reset_index()

#Analyze the correlation
correlation_new = club_stats_new[['hosting', 'is_win']].corr().iloc[0, 1]

# printing the correlation
print(f"Correlation between Hosting Games and Number of Wins: {correlation_new}")

#Visualization
plt.bar(club_stats_new['club_id'], club_stats_new['hosting'], label='Number of Home Games')
plt.bar(club_stats_new['club_id'], club_stats_new['is_win'], label='Number of Wins')
plt.xlabel('Club ID')
plt.ylabel('Count')
plt.title('Hosting Games vs. Number of Wins')
plt.legend()
plt.show()

```

```python
plt.figure(figsize=(20, 7))
plt.bar(club_stats_new['hosting'].astype(str) + ' Home Games', club_stats_new['is_win'])
plt.xlabel('Number of Home Games')
plt.ylabel('Average Number of Wins')
plt.title('Average Number of Wins for Different Numbers of Home Games')
plt.show()
```

```python
#third question: s there a relationship between a football club's squad characteristics such as the "squad size"and its performance in terms of "own goals" scored in games?

# calculate the average squad size and total own goals
club_stats_own_goals = grouped_data.agg({'squad_size': 'mean', 'own_goals': 'sum'}).reset_index()

# Analyze the correlation
correlation_own_goals = club_stats_own_goals[['squad_size', 'own_goals']].corr().iloc[0, 1]

# Display the correlation
print(f"Correlation between Squad Size and Total Own Goals: {correlation_own_goals}")

#Visualization
plt.scatter(club_stats_own_goals['squad_size'], club_stats_own_goals['own_goals'])
plt.xlabel('Average Squad Size of Club')
plt.ylabel('Total Own Goals Scored by Club')
plt.title('Relationship between Squad Size and Total Own Goals')
plt.show()
```

```python
# another visualization method: box plot to show the distribution of own goals based on squad size
plt.figure(figsize=(12, 6))
sns.boxplot(x=merged_data['squad_size'], y=merged_data['own_goals'])
plt.xlabel('Squad Size of Club')
plt.ylabel('Own Goals Scored by Club')
plt.title('Distribution of Own Goals based on Squad Size')
plt.show()
```

```python
#fourth question: Is there a relationship between a football club's percentage of foreign players and its performance when hosting games, measured by the number of wins?'


#calculate the average percentage of foreign players and total wins when hosting
grouped_data_foreigners = merged_data.groupby('club_id')
club_stats_foreigners = grouped_data_foreigners.agg({'foreigners_percentage': 'mean', 'is_win': 'sum'}).reset_index()

#calculating the correlation
correlation_foreigners = club_stats_foreigners[['foreigners_percentage', 'is_win']].corr().iloc[0, 1]
print(f"Correlation between Foreigners Percentage and Number of Wins when Hosting: {correlation_foreigners}")

#Visualization
plt.scatter(club_stats_foreigners['foreigners_percentage'], club_stats_foreigners['is_win'])
plt.xlabel('Foreigners Percentage of Club')
plt.ylabel('Number of Wins when Hosting')
plt.title('Relationship between Foreigners Percentage and Wins when Hosting')
plt.show()
```

```python
#another visualization method
merged_data['foreigners_percentage_range'] = pd.cut(merged_data['foreigners_percentage'], bins=[0, 20, 40, 60, 80, 100])
grouped_data_foreigners_range = merged_data.groupby('foreigners_percentage_range')
foreigners_range_stats = grouped_data_foreigners_range['is_win'].mean().reset_index()

plt.figure(figsize=(12, 6))
sns.barplot(x='foreigners_percentage_range', y='is_win', data=foreigners_range_stats)
plt.xlabel('Foreigners Percentage Range')
plt.ylabel('Average Number of Wins')
plt.title('Average Number of Wins for Different Ranges of Foreigners Percentage')
plt.show()
```

```python
#fifth question: does the average age has an impact on the number of game wins?

#calculating the average age and the total number of wins
grouped_data_age = merged_data.groupby('club_id')
club_stats_age = grouped_data_age.agg({'average_age': 'mean', 'is_win': 'sum'}).reset_index()

# Analyze the correlation between both features
correlation_age = club_stats_age[['average_age', 'is_win']].corr().iloc[0, 1]

# Display the correlation
print(f"Correlation between Average Squad Age and Number of Wins: {correlation_age}")

# Visualization
plt.scatter(club_stats_age['average_age'], club_stats_age['is_win'])
plt.xlabel('Average Squad Age of Club')
plt.ylabel('Number of Wins')
plt.title('Relationship between Average Squad Age and Number of Wins')
plt.show()
```

```python
# Creating a box plot to show the distribution of wins based on average squad age
plt.figure(figsize=(30, 6))
sns.boxplot(x=merged_data['average_age'], y=merged_data['is_win'])
plt.xlabel('Average Squad Age of Club')
plt.ylabel('Number of Wins')
plt.title('Distribution of Wins based on Average Squad Age')
plt.show()
```
